# SEABORN

## BARPLOTS

```python
import seaborn as sns
```

```python
url="https://raw.githubusercontent.com/mehakbhartiya/Sleep-Health-and-Lifestyle/refs/heads/main/Dataset/Sleep_health_and_lifestyle_dataset.csv"
sleep_ds = pd.read_csv(url)
sleep_ds
```

```python
import seaborn as sns
```

```python
sns.barplot(data=sleep_ds, x='Stress Level', y='Daily Steps', errorbar=None)
```

```python
fig, ax = plt.subplots(figsize=(14, 7.7))
sns.barplot(data=sleep_ds, x='Occupation', y='Daily Steps', estimator=sum, hue='Occupation',dodge =False ,errorbar=None,ax=ax)
```

```python
plt.figure(figsize=(8,5), tight_layout= False)
sns.barplot(data=sleep_ds, x='BMI Category', y='Physical Activity Level', errorbar=None, hue='BMI Category', dodge=False, estimator=sum)
plt.title('BMI Category V/S Physical Activity Level')
```

```python
fig,(ax1,ax2)=plt.subplots(1,2, figsize=(8,5))
sns.barplot(data=sleep_ds, x='Sleep Disorder', y='Age', errorbar=None, hue='Sleep Disorder', dodge=False, estimator=sum, ax=ax1)
sns.barplot(data=sleep_ds, x='Gender', y='Age', errorbar=None, hue='Gender', dodge=False, estimator=sum, ax=ax2)
```

## COUNTPLOT

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Gender',data=x)
```

## VIOLIN PLOT

```python
plt.figure(figsize=(15,5),tight_layout=False)
sns.violinplot(data=x, x='Sleep Disorder', y='Quality of Sleep',fliersize=0,dodge=False)
```

## STRIP PLOT

```python
plt.figure(figsize=(10,5),tight_layout=False)
sns.stripplot(data=x, x='Sleep Disorder', y='Quality of Sleep',dodge=False)
```

## POINT PLOT

```python
plt.figure(figsize=(15,5),tight_layout=False)
sns.pointplot(data=x, x='Sleep Disorder', y='Quality of Sleep',dodge=False)
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='Sleep Disorder',y='Quality of Sleep',data=sleep_ds,linestyle='--')
```
